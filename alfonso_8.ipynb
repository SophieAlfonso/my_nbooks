```python
import pandas as pd
import seaborn as sns
```

```python
avo_url = '/Users/sophiealfonso/Documents/my_nbooks/avocado.csv'
avocado = pd.read_csv(avo_url)
avocado.head()
```

```python
avocado.info()
```

```python
avocado['Date'] = avocado['Date'].apply(pd.to_datetime)
avocado.info()
```

```python
avo_new = avocado.query('year == 2015 and region == "Albany" and type == "conventional"')[['Date', 'Total Bags', 'Small Bags']]
avo_new.head()
```

```python
avo_melt = pd.melt(avocado, id_vars = 'Date', 
                   value_vars = ['Total Bags', 'Small Bags'],
                   var_name = 'Bags', value_name = 'Count')
avo_melt
```

```python
g = sns.relplot(data = avo_melt, kind = 'line', x = 'Date', y = 'Count', hue = 'Bags')
for ax in g.axes.flat:
    ax.tick_params('x', labelrotation = 30)
```

```python
avocado['Sales Volume'] = pd.qcut(avocado['Total Volume'], q = 4,
                        labels=['Poor','Modest','Good','Excellent']) 
avocado.head()
```

```python
sns.catplot(data = avocado, kind = 'count', x = 'year', hue = 'Sales Volume')
```

```python
avo_melt_2 = pd.melt(avocado, id_vars = 'year', 
                   value_vars = ['Small Bags', 'Large Bags', 'XLarge Bags'],
                   var_name = 'Bags', value_name = 'Count')
avo_melt_2
```

```python
h = sns.relplot(data = avo_melt_2, kind = 'line', x = 'year', y = 'Count', hue = 'Bags')
for ax in h.axes.flat:
    ax.tick_params('x', labelrotation = 30)
```
